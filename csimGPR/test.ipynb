```python
import matplotlib.pyplot as plt
import numpy as np

# 
vertices = np.array([[1, 1], [2, 3], [4, 2], [3, 1]])

# 
fig, ax = plt.subplots()

# 
polygon = plt.Polygon(vertices, closed=True, facecolor='lightblue')
ax.add_patch(polygon)

#  x  y 
ax.set_xlim(0, 5)
ax.set_ylim(0, 4)

# 

plt.savefig('polygon.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 
fig, ax = plt.subplots()

# fill_between
ax.fill_between(x, y, color='lightblue', hatch='?')

# 
ax.plot(x, y, label='sin(x)')

# 
ax.legend()

# 
plt.show()
```
