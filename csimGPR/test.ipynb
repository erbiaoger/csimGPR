```python
import matplotlib.pyplot as plt
import numpy as np

# 创建 x 值的数组
x = np.linspace(0, 10, 100)

# 创建两条线的 y 值数组
y1 = np.sin(x)
y2 = np.cos(x)

# 绘制两条线
plt.plot(x, y1)
plt.plot(x, y2)

# 填充两条线之间的区域
hatchs = ['/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*']
plt.fill_between(x, y1, y2, hatch='', facecolor='lightgreen', edgecolor='c', alpha=1, label='Line 1')
#plt.fill_between(x, y1, y2, hatch='o', facecolor='orange', edgecolor='c', alpha=0.5)
#plt.fill_between(x, y1, y2, hatch='*', facecolor='orange', edgecolor='c', alpha=0.5)
#plt.fill_between(x, y1, y2, hatch='.', facecolor='orange', edgecolor='c', alpha=0.5)


# 添加图例
plt.legend()

# 显示图形

plt.savefig('test.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建自定义的图案顶点坐标
vertices = np.array([[1, 1], [2, 3], [4, 2], [3, 1]])

# 创建一个新的图形对象
fig, ax = plt.subplots()

# 创建一个多边形对象，并添加到图形中
polygon = plt.Polygon(vertices, closed=True, facecolor='lightblue')
ax.add_patch(polygon)

# 设置 x 轴和 y 轴的范围
ax.set_xlim(0, 5)
ax.set_ylim(0, 4)

# 显示图形

plt.savefig('polygon.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建示例数据
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 创建图形对象和子图
fig, ax = plt.subplots()

# 使用fill_between函数填充颜色并添加问号填充效果
ax.fill_between(x, y, color='lightblue', hatch='?')

# 绘制曲线
ax.plot(x, y, label='sin(x)')

# 添加图例
ax.legend()

# 显示图形
plt.show()
```
